```
-- SQL query (we'll build the WHERE clause w/ Python!)
select n_name as country, count(c.*) as tot_customers
from snowflake_sample_data.tpch_sf1.nation n
join snowflake_sample_data.tpch_sf1.customer c
on n_nationkey = c_nationkey
-- where n_name IN ('CANADA', 'BRAZIL', 'CHINA')
group by n_name
order by n_name
```

```
from snowflake.snowpark.context import get_active_session

session = get_active_session()
conn = session.connection
qh = session.query_history()

query = """
select n_name as country, count(c.*) as tot_customers
from snowflake_sample_data.tpch_sf1.nation n
join snowflake_sample_data.tpch_sf1.customer c
on n_nationkey = c_nationkey
-- where n_name IN ('CANADA', 'BRAZIL', 'CHINA')
group by n_name
order by n_name
"""
```

```
# w/ Pandas dataframe, ~pd.read_sql(query, conn)
df = conn.cursor().execute(query).fetch_pandas_all()
df
```

```
# filter w/ Pandas DataFrame
df[df['COUNTRY'].isin(['CANADA', 'BRAZIL', 'CHINA'])]
```

```
# check generated SQL queries
qh.queries
```

```
# w/ Snowpark dataframe
df = session.sql(query)
df
```

```
# filter w/ Snowpark DataFrame
df.filter(df['COUNTRY'].isin('CANADA', 'BRAZIL', 'CHINA'))
```
